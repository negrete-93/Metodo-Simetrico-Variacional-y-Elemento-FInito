#Método Simétrico Variacional

Es una familia de enfoques que combinan principios variacionales con simetrías del problema para encontrar soluciones o aproximaciones óptimas. Se usa mucho en física, matemáticas aplicadas, mecánica, control, optimización y hasta en econometría/IO cuando el sistema tiene estructura.


Es simétrico porque el problema no cambia bajo ciertas transformaciones, entonces este método busca la mejor solución posible asegurando que respete la simetría natural del sistema.

El método variacional simétrico reduce el orden de derivación requerido, permite el uso de funciones base menos restrictivas y produce una estructura algebraica compatible con el método de los elementos finitos.



**Problema 5.5.13**
$$
\frac{d^2 u}{dx^2} + u = x, \quad 0 < x < 1
$$

sujeto a las condiciones de frontera:

$$
i) u(0) = 0, \quad u(1) = 0
$$


```python
#Importamos las librerías necesarias

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos la variable simbólica
x = sp.symbols('x')

#Ahora definimos nuestras betas

def beta(i):
    return (x**i) * (1 - x)

def beta_prima(i):
    return sp.diff(beta(i), x)
```

Construimos nuestra matriz de rigidez

```python
def Matriz_Rigidez(N):

    K = sp.zeros(N, N)

    for i in range(1, N + 1):

        for j in range(1, N + 1):
            integrand = -beta_prima(i) * beta_prima(j) + beta(i) * beta(j)
            K[i-1, j-1] = sp.integrate(integrand, (x, 0, 1))

    return K
```

Construimos nuestro vector de carga

```python
def load_vector(N):

    F = sp.zeros(N, 1)

    for i in range(1, N + 1):
        F[i-1, 0] = sp.integrate(x * beta(i), (x, 0, 1))

    return F
```

Resolvemos el problema variacional

```python
def solve_variational(N):

    K = Matriz_Rigidez(N)
    F = load_vector(N)

    alpha = K.LUsolve(F)
    alpha_np = np.array(alpha, dtype=float).flatten()
    return alpha_np
```


Buscamos la solución aproximada:
$$
u_N(x) = \sum_{i=1}^{N} \alpha_i \, \beta_i(x)
$$

```python
def uh_symbolico(alpha):
    N = len(alpha)
    u_h = 0
    for i in range(1, N + 1):
        u_h += alpha[i-1] * beta(i)
    return sp.simplify(u_h)
```

Definimos el orden del método y  obtenemos:



```python
#Definimos nuestro espacio Vh
N = 6
alpha = solve_variational(N)

#Construimos la solución aproximada
u_h = uh_symbolico(alpha)
u_h_fun = sp.lambdify(x, u_h, "numpy")

# Solución exacta
u_exact = x - sp.sin(x) / sp.sin(1)
u_exact_fun = sp.lambdify(x, u_exact, "numpy")

#Construimos la malla
xx = np.linspace(0, 1, 400)
yh = u_h_fun(xx)
ye = u_exact_fun(xx)

# Calculamos el error L2 aproximado

err_L2 = np.sqrt(np.trapezoid((ye - yh)**2, xx))

print(f"N = {N}\n")
print("alpha =", alpha,"\n")
print("u_h(x) =", u_h,"\n")
print(f"Error L2 (aprox) = {err_L2:.6e}")
```

Mostramos la gráfica

```python
# Gráfica
plt.figure()
plt.plot(xx, ye, label="Exacta")
plt.plot(xx, yh, "--", label=f"Aprox (Ritz) N={N}")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.legend()
plt.show()
```

**Problema 5.5.13**

Resolvemos el segundo inciso del mism problema

$$
\frac{d^2 u}{dx^2} + u = x, \quad 0 < x < 1
$$

sujeto a las condiciones de frontera:

$$
ii)u(0) = 0, \quad u'(1) = 1
$$


```python
#Importamos las librerías necesarias
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos la variable simbólica
x = sp.symbols('x')

#Ahora definimos nuestras betas
def beta(i):
    return x**i

def beta_prima(i):
    return sp.diff(beta(i), x)
```

Construimos nuestra matriz de rigidez

```python
def Matriz_Rigidez(N):
    K = sp.zeros(N, N)
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            integrand = -beta_prima(i) * beta_prima(j) + beta(i) * beta(j)
            K[i-1, j-1] = sp.integrate(integrand, (x, 0, 1))
    return K
```

Construimos nuestro vector de carga

```python
def load_vector(N):
    F = sp.zeros(N, 1)
    for i in range(1, N + 1):
        F[i-1, 0] = sp.integrate(x * beta(i), (x, 0, 1)) - beta(i).subs(x, 1)
    return F
```

Resolvemos el problema variacional

```python
def solve_variational(N):
    K = Matriz_Rigidez(N)
    F = load_vector(N)
    alpha = K.LUsolve(F)
    alpha_np = np.array(alpha, dtype=float).flatten()
    return alpha_np
```


Buscamos la solución aproximada:
$$
u_N(x) = \sum_{i=1}^{N} \alpha_i \, \beta_i(x)
$$

```python
def uh_symbolico(alpha):
    N = len(alpha)
    u_h = 0
    for i in range(1, N + 1):
        u_h += alpha[i-1] * beta(i)
    return sp.simplify(u_h)
```

Obtenemos:

```python
N = 6
alpha = solve_variational(N)

u_h = uh_symbolico(alpha)
u_h_fun = sp.lambdify(x, u_h, "numpy")

u_exact = x
u_exact_fun = sp.lambdify(x, u_exact, "numpy")

xx = np.linspace(0, 1, 400)
yh = u_h_fun(xx)
ye = u_exact_fun(xx)

err_L2 = np.sqrt(np.trapezoid((ye - yh)**2, xx))

print(f"N = {N}\n")
print("alpha =", alpha, "\n")
print("u_h(x) =", u_h, "\n")
print(f"Error L2 (aprox) = {err_L2:.6e}")
```

Graficamos

```python
plt.figure()
plt.plot(xx, ye, label="Exacta (u=x)")
plt.plot(xx, yh, "--", label=f"Aprox (Ritz) N={N}")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.legend()
plt.show()
```

#Elemento Finito

Divide el dominio continuo en subdominios pequeños (elementos) y aproxima la solución de la EDP en cada uno usando funciones simples.

No resuelve la EDP “tal cual”, sino una versión aproximada construida a partir de:


*   Una formulación débil (variacional)
*   Un espacio finito dimensional generado por los elementos.

El método del elemento finito hace esto:


$$
V -> V_h \subset V
$$

donde:

*   $V_h$ es un espacio finito dimensional
*   Y es construido con funciones asociadas a los elementos finitos


Entonces el problema se vuelve:

Encontrar $u_h \in V_h$, tal que:
$$
a(u_h,v_h)=l(v_h) \forall v_h \in V_h
$$
El elemento finito es la base concreta del método variacional.



**Problema 5.6.9**
$$
-u''+u+x
$$

sujeto a:

$$
 u(0) = u(1) = 0
$$

La formulación débil es:
$$
∫_{0}^{1}u´v´dx + ∫_{0}^{1}uvdx = ∫_{0}^{1}xv dx
$$

y con FEM se arma:

$(K+M)U=F$

donde:


*   $K:$ rigidez $Φ'_i Φ'_j$
*   $M:$ masa $Φ_i Φ_j$
*   $F:$ carga $x Φ_i$





```python
#Importamos librerías
import numpy as np
import matplotlib.pyplot as plt
```

Resolvemos numéricamente la EDP

Usando elementos finitos lineales (P1) sobre una malla uniforme.

Construimos la matriz de rigidez local:

Y la matriz de masa local:image.png

```python
def fem_1d_poisson_reaction(num_elems: int):

    # Definimos la malla
    a, b = 0.0, 1.0
    nodes = np.linspace(a, b, num_elems + 1)
    h = nodes[1] - nodes[0]
    n_nodes = len(nodes)

    # Matriz global y vector global
    A = np.zeros((n_nodes, n_nodes))  # A = K + M
    F = np.zeros(n_nodes)

    # Matrices locales
    Ke = (1.0 / h) * np.array([[1.0, -1.0],
                               [-1.0, 1.0]])

    # Construimos la matriz de rigidez local
    Me = (h / 6.0) * np.array([[2.0, 1.0],
                               [1.0, 2.0]])

    # Ensamblamos por elemento
    for e in range(num_elems):
        xL, xR = nodes[e], nodes[e + 1]

        # Índices globales de los 2 nodos del elemento
        idx = [e, e + 1]

        # Con P1, se puede integrar con cuadratura de Gauss de 2 puntos

        gauss_pts = np.array([-1.0 / np.sqrt(3.0), 1.0 / np.sqrt(3.0)])
        gauss_wts = np.array([1.0, 1.0])

        Fe = np.zeros(2)
        for (xi, w) in zip(gauss_pts, gauss_wts):
            # Mapeo de referencia [-1,1] -> [xL,xR]
            x = (xL + xR) / 2.0 + (h / 2.0) * xi
            J = h / 2.0

            # Funciones de forma P1 en el elemento de referencia
            phi1 = (1.0 - xi) / 2.0
            phi2 = (1.0 + xi) / 2.0

            Fe[0] += w * (x * phi1) * J
            Fe[1] += w * (x * phi2) * J

        # Ensamblamos A y F
        Ae = Ke + Me
        for i_local in range(2):
            I = idx[i_local]
            F[I] += Fe[i_local]
            for j_local in range(2):
                Jg = idx[j_local]
                A[I, Jg] += Ae[i_local, j_local]

    # Tomamos solo incógnitas interiores
    interior = np.arange(1, n_nodes - 1)
    Aii = A[np.ix_(interior, interior)]
    Fi = F[interior]

    # Resolvemos
    Ui = np.linalg.solve(Aii, Fi)

    # Reconstruimos la solución completa
    U = np.zeros(n_nodes)
    U[interior] = Ui
    U[0] = 0.0
    U[-1] = 0.0

    return nodes, U
```

```python
ne = 40  # número de elementos
x, uh = fem_1d_poisson_reaction(ne)

# Solución exacta
u_exact = x - np.sinh(x) / np.sinh(1.0)
# Error L2 aproximado
err_L2 = np.sqrt(np.trapezoid((u_exact - uh) ** 2, x))
print(f"Elementos = {ne}")
print(f"Error L2 (aprox) = {err_L2:.6e}")
```

```python
plt.figure()
plt.plot(x, u_exact, label="Exacta")
plt.plot(x, uh, "--", label=f"FEM P1 (ne={ne})")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.legend()
plt.show()
```
